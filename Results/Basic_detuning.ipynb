```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from simulation import Simulation
```

## Direct SC

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_mod/'))
beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0
sim.build_particles(n_particles=500, nemitt_x_0=beta_gamma*60e-6, nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 500, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100)
sim.run()
sim.save('data/basic/dir_sc')
```

## Indirect SC

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_mod/'))
beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0
sim.build_particles(n_particles=500, nemitt_x_0=beta_gamma*60e-6, nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 500, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100, indirect_spacecharge=True)
sim.run()
sim.save('data/basic/ind_sc')
```
