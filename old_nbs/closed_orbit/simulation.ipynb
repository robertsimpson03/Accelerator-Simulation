# Detuning as a function of offset

```python
#offset_list = [2e-4, 5e-4, 1e-3, 2e-3, 5e-3]
offset_list = [1e-4, 1e-2, 2e-2, 5e-2]
for offset in offset_list:
    sim = Simulation(
        threads=0,
        folder=Path('../Lattice_Files/02_Aperture_Lattice/')
    )
    sim.build_particles(co=True, offset_x=offset, offset_y=offset, nemitt_x_0=1e-6, nemitt_y_0=1e-6)
    sim.set_up(n_turns = 1000, n_monitors = 10,)
    sim.run(spacecharge=True, beam_intensity=2e11, n_interactions=100, xfields=False, indirect_spacecharge=False)
    sim.save(f'sweep_offset/offset_{offset}')
```
