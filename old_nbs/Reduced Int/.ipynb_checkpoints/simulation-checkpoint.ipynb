# Smaller intensity- indirect

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_mod/'))
beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0

sim.build_particles(n_particles=1000, 
                    nemitt_x_0=beta_gamma*60e-6,
                    nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 500, n_monitors = 10)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100, indirect_spacecharge=True)
sim.run()
sim.save('data/indirect_sc')
```

# Estimate shift

```python
from IPython.display import display, Math
from scipy.constants import e, epsilon_0, m_p, c
ke = 1/(4*np.pi*epsilon_0)
R = sim.line.get_length()  /(2*np.pi)
Qx = sim.twiss.qx
Qy = sim.twiss.qy

beam_intensity = 5e12
bunch_length  = 60#a*c/f
_lambda = beam_intensity/bunch_length
beta = sim.line.particle_ref.beta0
gamma = sim.line.particle_ref.gamma0
p0c = sim.line.particle_ref.p0c

nemitt_x = beta*gamma*60e-6
nemitt_y = beta*gamma*60e-6

betx = np.mean(sim.twiss.betx)
bety = np.mean(sim.twiss.bety)

sigma_x = np.sqrt(nemitt_x*betx)
sigma_y = np.sqrt(nemitt_y*bety)

K_sc = 2*ke*e*_lambda/(beta*gamma**2*p0c)

dQx = K_sc * R**2 / (1*sigma_x*(sigma_x+sigma_y)*Qx)
dQy = K_sc * R**2 / (1*sigma_y*(sigma_x+sigma_y)*Qy)

display(Math(rf"$\Delta Q_x = {dQx[0]}$"))
display(Math(rf"$\Delta Q_y = {dQy[0]}$"))

print(betx)
print(bety)

coasting_int = 2*np.pi*R*beam_intensity/bunch_length
print(coasting_int*1e-14)
```
