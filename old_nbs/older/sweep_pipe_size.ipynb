```python
import numpy as np
from simulation import Simulation
```

```python
sim = Simulation(threads=10)
s_list = np.linspace(0, sim.length, 26, endpoint=False)
strength_list = 1e-4 * np.cos(13 * 2*np.pi * s_list / sim.line.get_length())
sim.add_dipole(strength=strength_list, s=s_list, mode='skew')
sim.set_up(n_particles = int(1e3),
          n_turns = 50,
          n_monitors = 100,
          nemitt_x_0=1e-6,
          nemitt_y_0=1e-6)
sim.run(spacecharge=True, beam_intensity=5e10, n_interactions=100, indirect_spacecharge=False)
sim.save('sweep_pipe_size_3/no_pipe')
```

```python
scale_factors = [1, 0.5, 0.2, 0.1, 0.05]
for scale_factor in scale_factors:
    sim = Simulation(threads=10)
    s_list = np.linspace(0, sim.length, 26, endpoint=False)
    strength_list = 1e-4 * np.cos(13 * 2*np.pi * s_list / sim.line.get_length())
    sim.add_dipole(strength=strength_list, s=s_list, mode='skew')
    sim.set_up(n_particles = int(1e3),
              n_turns = 50,
              n_monitors = 100,
              nemitt_x_0=1e-6,
              nemitt_y_0=1e-6)
    sim.run(spacecharge=True, beam_intensity=5e10, n_interactions=100, indirect_spacecharge=True,
           pipe_enlargement=scale_factor)
    sim.save(f'sweep_pipe_size_3/sf:{scale_factor}')
```
