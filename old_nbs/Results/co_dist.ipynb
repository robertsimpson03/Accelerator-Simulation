```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from simulation import Simulation
```

```python
x = np.linspace(0, 0.05, 500)
p = np.zeros_like(x)
y = np.full_like(a=x, fill_value=0.005)


sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
nemitt = sim.line.particle_ref.beta0 * sim.line.particle_ref.gamma0 * 60e-6
sim.particles = sim.line.build_particles(
    x=x+sim.line.twiss(method='4d', freeze_longitudinal=True).x[0],
    y=y+sim.line.twiss(method='4d', freeze_longitudinal=True).y[0],
    px=p+sim.line.twiss(method='4d', freeze_longitudinal=True).px[0],
    py=p+sim.line.twiss(method='4d', freeze_longitudinal=True).py[0],
                         method='4d', freeze_longitudinal=True)
sim.n_particles=len(x)
sim.nemitt_x_0=nemitt
sim.nemitt_y_0=nemitt
sim.set_up(n_turns = 100, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100,
                    nemitt_x=nemitt, nemitt_y=nemitt)
sim.run()
sim.save('data/co_dist/dir_y')


sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
s_list = np.linspace(0, sim.length, 26, endpoint=False)
strength_list = 1e-3 * np.cos(13 * 2*np.pi * s_list / sim.line.get_length())
sim.add_dipole(strength=strength_list, s=s_list, mode='normal')
sim.particles = sim.line.build_particles(
    x=x+sim.line.twiss(method='4d', freeze_longitudinal=True).x[0],
    y=y+sim.line.twiss(method='4d', freeze_longitudinal=True).y[0],
    px=p+sim.line.twiss(method='4d', freeze_longitudinal=True).px[0],
    py=p+sim.line.twiss(method='4d', freeze_longitudinal=True).py[0],
                         method='4d', freeze_longitudinal=True)
sim.nemitt_x_0=nemitt
sim.nemitt_y_0=nemitt
sim.n_particles=len(x)
sim.set_up(n_turns = 100, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100,
                    nemitt_x=nemitt, nemitt_y=nemitt, indirect_spacecharge=False)
sim.run()
sim.save('data/co_dist/dir_distorted_y')
```
