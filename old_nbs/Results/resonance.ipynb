```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from simulation import Simulation
```

## Direct SC

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))

s_list = np.linspace(0, sim.length, 26, endpoint=False)
strength_list = 1e-3 * np.cos(13 * 2*np.pi * s_list / sim.line.get_length())
sim.add_dipole(strength=strength_list, s=s_list, mode='skew') # normal

beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0
sim.build_particles(n_particles=1000, nemitt_x_0=beta_gamma*60e-6, nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 100, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100)
sim.run()
sim.save('data/resonance/dir_sc_kick_skew')
```

## Indirect SC

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0

s_list = np.linspace(0, sim.length, 26, endpoint=False)
strength_list = 1e-3 * np.cos(13 * 2*np.pi * s_list / sim.line.get_length())
sim.add_dipole(strength=strength_list, s=s_list, mode='skew') # normal

sim.build_particles(n_particles=1000, nemitt_x_0=beta_gamma*60e-6, nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 100, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100, indirect_spacecharge=True)
sim.run()
sim.save('data/resonance/ind_sc_kick_skew')
```
