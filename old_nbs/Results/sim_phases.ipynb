```python
import sys
from pathlib import Path
import xpart as xp

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from simulation import Simulation
```

```python
sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice_mod/'))
beta_gamma = sim.line.particle_ref.beta0*sim.line.particle_ref.gamma0
```

```python
nemitt_x = 10e-6
gemitt_x = nemitt_x / beta_gamma

num_particles = 32
phases = np.linspace(0, 2*np.pi, num_particles, endpoint=False)

beta_x = sim.line.twiss(method='4d', freeze_longitudinal=True).betx[0]
alpha_x = sim.line.twiss(method='4d', freeze_longitudinal=True).alfx[0]

x = np.sqrt(gemitt_x * beta_x) * np.cos(phases)
px = -np.sqrt(gemitt_x / beta_x) * (np.sin(phases) + alpha_x * np.cos(phases))

sim.particles = xp.build_particles(particle_ref=sim.line.particle_ref, x=x, px=px)
sim.n_particles = num_particles
```

```python
sim.set_up(n_turns = 100, n_monitors = 100)
sim.add_spacecharge(beam_intensity=1e13, n_interactions=100, 
                    nemitt_x=beta_gamma*60e-6, nemitt_y=beta_gamma*60e-6,)
sim.nemitt_x_0 = beta_gamma*60e-6
sim.nemitt_y_0 = beta_gamma*60e-6
sim.run()
sim.save('data/phases/dir_sc')
```
