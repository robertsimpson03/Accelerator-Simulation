```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

import simulation_functional as sim
import xobjects as xo
import copy
```

```python
gemitt = 60e-6
n_turns=5000
n_particles = 250

gamx = 0.13112848113527537 # gamma at 0
gamy = 0.35349943817994744

sqrt_Jx_max = np.sqrt(5*gemitt) #gamx * 0.05**2
sqrt_Jy = np.sqrt(0.01*gemitt)
sqrt_Jx = np.linspace(0, sqrt_Jx_max, 200)
```

```python
print(sqrt_Jx)
```

```python
for kick in [False, True]:
    for indirect in [False, True]:
        line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))

        if kick==True:
            s_list = np.linspace(0, line.get_length(), 13*2, endpoint=False)
            strength_list = 5e-3 * np.cos(13 * 2*np.pi * s_list / line.get_length())
            line = sim.add_dipole(line, strength=strength_list, s=s_list, mode='normal')

        gamx = line.twiss(method='4d', freeze_longitudinal=True).gamx[0]
        gamy = line.twiss(method='4d', freeze_longitudinal=True).gamy[0]
        x = sqrt_Jx * np.sqrt(1/gamx)
        y = sqrt_Jy * np.sqrt(1/gamy)
        p = np.zeros_like(x)
        
        particles = line.build_particles(
            x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
            y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
            px=p+line.twiss(method='4d', freeze_longitudinal=True).px[0],
            py=p+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                                 method='4d', freeze_longitudinal=True)
        
        line = sim.add_monitors(line, particles, n_monitors=1, n_turns=n_turns)
        
        twiss = line.twiss(method='4d', freeze_longitudinal=True)
        
        line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=1e13,
                                                     n_interactions=100, 
                                                     gemitt=gemitt, indirect=indirect)
        
        line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
        line.track(particles, num_turns=n_turns, time=True)
        
        sim.save(line, particles, twiss, filename=f'data/kick_{kick}_ind_{indirect}', aper_df=aper_df, beam_df=beam_df)
```

```python
Jx = 0.01314**2
for kick in [False, True]:
    for indirect in [False, True]:
        line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))

        if kick==True:
            s_list = np.linspace(0, line.get_length(), 13*2, endpoint=False)
            strength_list = 5e-3 * np.cos(13 * 2*np.pi * s_list / line.get_length())
            line = sim.add_dipole(line, strength=strength_list, s=s_list, mode='normal')

        gamx = line.twiss(method='4d', freeze_longitudinal=True).gamx[0]
        gamy = line.twiss(method='4d', freeze_longitudinal=True).gamy[0]
        betx = line.twiss(method='4d', freeze_longitudinal=True).betx[0]
        bety = line.twiss(method='4d', freeze_longitudinal=True).bety[0]
        alfx = line.twiss(method='4d', freeze_longitudinal=True).alfx[0]
        alfy = line.twiss(method='4d', freeze_longitudinal=True).alfy[0]
        
        phases = np.linspace(0, 2*np.pi, n_particles, endpoint=False)
        x_norm = np.sqrt(Jx) * np.cos(phases)
        px_norm = -np.sqrt(Jx) * np.sin(phases)
        x = x_norm * np.sqrt(betx)
        px = (px_norm - alfx * x_norm) / np.sqrt(betx)
        
        
        y = sqrt_Jy / np.sqrt(gamy)
        py = 0
        
        particles = line.build_particles(
            x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
            y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
            px=px+line.twiss(method='4d', freeze_longitudinal=True).px[0],
            py=py+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                                 method='4d', freeze_longitudinal=True)
        
        line = sim.add_monitors(line, particles, n_monitors=1, n_turns=n_turns)
        
        twiss = line.twiss(method='4d', freeze_longitudinal=True)
        
        line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=1e13,
                                                     n_interactions=100, 
                                                     gemitt=gemitt, indirect=indirect)
        
        line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
        line.track(particles, num_turns=n_turns, time=True)
        
        sim.save(line, particles, twiss, filename=f'data/kick_{kick}_ind_{indirect}_on_res', aper_df=aper_df, beam_df=beam_df)
```
