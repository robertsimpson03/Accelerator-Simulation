```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')

import scienceplots

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('science')
import analysis_functional as a
```

```python
Qxs = []
Qys = []
qxs = []
qys = []
survivor_masks = []
turn=10
for angle in ['normal', 'skew']:
    for ind in [False, True]:
        res = a.load_simulation(f'data/{angle}_ind_{ind}.h5')
        Qx, Qy = a.get_tunes(res.x, res.y, res.px, res.py, res.twiss_df_at_mon, res.survivor_mask)
        Qxs.append(Qx)
        Qys.append(Qy)
        qxs.append(res.qx)
        qys.append(res.qy)
        print(np.sum(res.survivor_mask))
        survivor_masks.append(res.survivor_mask)
        
        fig, ax = plt.subplots(figsize=(10,4))
        ax.plot(res.aper_df['s'], res.aper_df['x'].values + res.aper_df['Lx'].values, c='k', label='Beam Pipe')
        ax.plot(res.aper_df['s'], res.aper_df['x'].values - res.aper_df['Lx'].values, c='k')
        ax.plot(res.aper_df['s'], res.beam_df['x0'].values + res.beam_df['sigma_x'].values, c='r', ls='--', label=r'$1\sigma$ (from Twiss)')
        ax.plot(res.aper_df['s'], res.beam_df['x0'].values - res.beam_df['sigma_x'].values, c='r', ls='--',)
        ax.plot(res.aper_df['s'], res.beam_df['x0'].values, c='r', label='Closed orbit')
        x = res.x.T[100*turn:100*(turn+1)]
        centroid_x = np.mean(x, axis=1)
        std_x = np.std(x, axis=1)
        ax.plot(res.twiss_df_at_mon['s'], centroid_x, c='b', label='Ensemble centroid')
        ax.plot(res.twiss_df_at_mon['s'], centroid_x+std_x, c='b', ls='--', label=r'Ensemble $1\sigma$')
        ax.plot(res.twiss_df_at_mon['s'], centroid_x-std_x, c='b', ls='--')
        ax.set_xlabel('$s$ / m')
        ax.set_ylabel('$x$ / m')
        fig.suptitle(f'X: {angle} - ind={ind} - Loss={(1000-np.sum(res.survivor_mask))/10}\%')
        plt.show()


        fig, ax = plt.subplots(figsize=(10,4))
        ax.plot(res.aper_df['s'], res.aper_df['y'].values + res.aper_df['Ly'].values, c='k', label='Beam Pipe')
        ax.plot(res.aper_df['s'], res.aper_df['y'].values - res.aper_df['Ly'].values, c='k')
        ax.plot(res.aper_df['s'], res.beam_df['y0'].values + res.beam_df['sigma_y'].values, c='r', ls='--', label=r'$1\sigma$ (from Twiss)')
        ax.plot(res.aper_df['s'], res.beam_df['y0'].values - res.beam_df['sigma_y'].values, c='r', ls='--',)
        ax.plot(res.aper_df['s'], res.beam_df['y0'].values, c='r', label='Closed orbit')
        y = res.y.T[100*turn:100*(turn+1)]
        centroid_y = np.mean(y, axis=1)
        std_y = np.std(y, axis=1)
        ax.plot(res.twiss_df_at_mon['s'], centroid_y, c='b', label='Ensemble centroid')
        ax.plot(res.twiss_df_at_mon['s'], centroid_y+std_y, c='b', ls='--', label=r'Ensemble $1\sigma$')
        ax.plot(res.twiss_df_at_mon['s'], centroid_y-std_y, c='b', ls='--')
        ax.set_xlabel('$s$ / m')
        ax.set_ylabel('$x$ / m')
        fig.suptitle(f'Y: {angle} - ind={ind} - Loss={(1000-np.sum(res.survivor_mask))/10}\%')
        plt.show()
```

```python
sF = a.load_simulation(f'data/skew_ind_False.h5')
sT = a.load_simulation(f'data/skew_ind_True.h5')
nF = a.load_simulation(f'data/normal_ind_False.h5')
nT = a.load_simulation(f'data/normal_ind_True.h5')
fig, axs = plt.subplots(2, 2, figsize=(12,6))

for particle in range(0, 1000, 100):
    axs[0,0].scatter(sF.x[particle][::100], sF.px[particle][::100], marker='.', s=0.5, alpha=0.9)
    axs[1,0].scatter(sT.x[particle][::100], sT.px[particle][::100], marker='.', s=0.5, alpha=0.5)
    axs[0,1].scatter(nF.x[particle][::100], nF.px[particle][::100], marker='.', s=0.5, alpha=0.9)
    axs[1,1].scatter(nT.x[particle][::100], nT.px[particle][::100], marker='.', s=0.5, alpha=0.9)

    axs[0,0].set_title('Direct - Skew')
    axs[1,0].set_title('Direct - Skew')
    axs[0,1].set_title('Indirect - Normal')
    axs[1,1].set_title('Indirect - Normal')
```
