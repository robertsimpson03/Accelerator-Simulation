# Default params with no SC, Direct and Indirect

```python
beta_gamma = 0.3934 

sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice/'))
sim.build_particles(n_particles=10000, 
                    nemitt_x_0=beta_gamma*60e-6, 
                    nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 100, n_monitors = 10)
sim.run()
sim.save('data/default/no_sc')

sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice/'))
sim.build_particles(n_particles=10000, 
                    nemitt_x_0=beta_gamma*60e-6, 
                    nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 100, n_monitors = 10)
sim.add_spacecharge(beam_intensity=5e13, n_interactions=100, 
                    indirect_spacecharge=False)
sim.run()
sim.save('data/default/dir_sc')

sim = Simulation(threads=10, folder=Path('../Lattice_Files/02_Aperture_Lattice/'))
sim.build_particles(n_particles=10000, 
                    nemitt_x_0=beta_gamma*60e-6, 
                    nemitt_y_0=beta_gamma*60e-6)
sim.set_up(n_turns = 100, n_monitors = 10)
sim.add_spacecharge(beam_intensity=5e13, n_interactions=100, 
                    indirect_spacecharge=True)
sim.run()
sim.save('data/default/ind_sc')
```
