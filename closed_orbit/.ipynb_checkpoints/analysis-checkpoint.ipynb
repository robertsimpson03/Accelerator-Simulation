# Detuning as a function of offset

```python
offset_list = [1e-4, 2e-4, 5e-4, 1e-3, 2e-3, 5e-3, 1e-2, 2e-2,]
DQx_list = []
DQy_list = []
Qx_list = []
Qy_list = []
for offset in offset_list:
    print('#######################################################')
    print(f'Offset = {offset}')
    print('#######################################################')
    analysis = Analysis(f'sweep_offset/offset_{offset}')
    analysis.tune_histogram(x_lims=(4.24,4.32), y_lims=(3.70,3.78))
    print(f'Tune from twiss: Qx={analysis.qx}, Qy={analysis.qy}')
    print(f'Tune of test particle: Qx={analysis.Qx[0]}, Qy={analysis.Qy[0]}')
    DQx_list.append(analysis.DQx)
    DQy_list.append(analysis.DQy)
    Qx_list.append(analysis.Qx)
    Qy_list.append(analysis.Qy)
```

```python
err_sigx = np.std(analysis.sigma_x)
sigx =  np.mean(analysis.sigma_x)
shiftx = DQx_list[0]
err_shiftx = shiftx*err_sigx/sigx

plt.plot(offset_list, DQx_list, label='Theoretical maxium')
plt.plot(offset_list, DQx_list+err_shiftx, ls='--', c='tab:blue')
plt.plot(offset_list, DQx_list-err_shiftx, ls='--', c='tab:blue')
plt.plot(offset_list, analysis.qx-Qx_list, label='Observed')
plt.xlabel('Beam Offset')
plt.ylabel('Tune shift (x)')
plt.legend()
plt.show()


err_sigy = np.std(analysis.sigma_y)
sigy =  np.mean(analysis.sigma_y)
shifty = DQy_list[0]
err_shifty = shifty*err_sigy/sigy

plt.plot(offset_list, DQy_list, label='Theoretical maxium')
plt.plot(offset_list, DQy_list+err_shifty, ls='--', c='tab:blue')
plt.plot(offset_list, DQy_list-err_shifty, ls='--', c='tab:blue')
plt.plot(offset_list, analysis.qy-Qy_list, label='Observed')
plt.xlabel('Beam Offset')
plt.ylabel('Tune shift (y)')
plt.legend()
plt.show()
```
