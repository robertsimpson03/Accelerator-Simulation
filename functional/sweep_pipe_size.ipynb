```python
import sys

from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

import simulation_functional as sim
import xobjects as xo
```

```python
n_turns = 100
gamx = 0.13112848113527537 # gamma at 0
Jx_max = gamx * 0.05**2
Jx = np.linspace(0, Jx_max, 100)
x = np.sqrt(Jx / gamx)
p = np.zeros_like(x)
y = np.full_like(a=x, fill_value=0.0001)


# No beam pipe
line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
particles = line.build_particles(
    x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
    y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
    px=p+line.twiss(method='4d', freeze_longitudinal=True).px[0],
    py=p+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                         method='4d', freeze_longitudinal=True)

line = sim.add_monitors(line, particles, n_monitors=100, n_turns=n_turns)
twiss = line.twiss(method='4d', freeze_longitudinal=True)
line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=5e12,
                                             n_interactions=100, 
                                             gemitt=60e-6, indirect=False)
line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
line.track(particles, num_turns=n_turns, time=True)
sim.save(line, particles, twiss, 
         filename='data/sweep_size_offset/none', 
         aper_df=aper_df, beam_df=beam_df)


#enlargements = [0.25, 0.5, 1, 1.5, 2, 3, 5]
enlargements = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.5, 2]
for en in enlargements:
    line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
    particles = line.build_particles(
        x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
        y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
        px=p+line.twiss(method='4d', freeze_longitudinal=True).px[0],
        py=p+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                             method='4d', freeze_longitudinal=True)
    
    line = sim.add_monitors(line, particles, n_monitors=100, n_turns=n_turns)
    twiss = line.twiss(method='4d', freeze_longitudinal=True)
    line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=5e12,
                                                 n_interactions=100, 
                                                 gemitt=60e-6, indirect=True, 
                                                 pipe_enlargement=en)
    line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
    line.track(particles, num_turns=n_turns, time=True)

    sim.save(line, particles, twiss, 
         filename=f'data/sweep_size_offset/{en}', 
         aper_df=aper_df, beam_df=beam_df)
```

```python
n_turns = 100

gamx = 0.13112848113527537 # gamma at 0
gamy = 0.35660305783564805

Jx_max = gamx * 0.02**2
Jy_max = gamy * 0.02**2
Jx = np.linspace(0, Jx_max, 20)
Jy = np.linspace(0, Jy_max, 20)
x = np.sqrt(Jx / gamx)
y = np.sqrt(Jy / gamy)
x, y = np.meshgrid(x,y)
x = x.flatten()
y = y.flatten()
p = np.zeros_like(x)

# No beam pipe
line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'), )
particles = line.build_particles(
    x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
    y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
    px=p+line.twiss(method='4d', freeze_longitudinal=True).px[0],
    py=p+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                         method='4d', freeze_longitudinal=True)

line = sim.add_monitors(line, particles, n_monitors=20, n_turns=n_turns)
twiss = line.twiss(method='4d', freeze_longitudinal=True)
line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=5e12,
                                             n_interactions=100, 
                                             gemitt=60e-6, indirect=False)
line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
line.track(particles, num_turns=n_turns, time=True)
sim.save(line, particles, twiss, 
         filename='data/sweep_2d/none', 
         aper_df=aper_df, beam_df=beam_df)


enlargements = [0.5, 1, 2]
for en in enlargements:
    line = sim.build_line(folder=Path('../Lattice_Files/02_Aperture_Lattice_res/'))
    particles = line.build_particles(
        x=x+line.twiss(method='4d', freeze_longitudinal=True).x[0],
        y=y+line.twiss(method='4d', freeze_longitudinal=True).y[0],
        px=p+line.twiss(method='4d', freeze_longitudinal=True).px[0],
        py=p+line.twiss(method='4d', freeze_longitudinal=True).py[0],
                             method='4d', freeze_longitudinal=True)
    
    line = sim.add_monitors(line, particles, n_monitors=20, n_turns=n_turns)
    twiss = line.twiss(method='4d', freeze_longitudinal=True)
    line, beam_df, aper_df = sim.add_spacecharge(line, beam_intensity=5e12,
                                                 n_interactions=100, 
                                                 gemitt=60e-6, indirect=True, 
                                                 pipe_enlargement=en)
    line.build_tracker(_context=xo.ContextCpu(omp_num_threads=10))
    line.track(particles, num_turns=n_turns, time=True)

    sim.save(line, particles, twiss, 
         filename=f'data/sweep_2d/{en}', 
         aper_df=aper_df, beam_df=beam_df)
```
