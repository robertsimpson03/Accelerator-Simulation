```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')
import numpy as np

from simulation import Simulation
```

```python
sim = Simulation(
        threads=10,
        folder=Path('../Lattice_Files/02_Aperture_Lattice_mod/')
    )
sim.build_particles(n_particles=1000, nemitt_x_0=1e-5, nemitt_y_0=1e-5)
sim.set_up(n_turns = 500, n_monitors = 10)
sim.run()
sim.save('data/no_sc')
```
