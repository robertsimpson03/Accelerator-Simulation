```python
import sys
from pathlib import Path

parent_folder = Path.cwd().parent
if str(parent_folder) not in sys.path:
    sys.path.append(str(parent_folder))
    
import warnings
warnings.filterwarnings('ignore')

import numpy as np
from simulation import Simulation
import xtrack as xt
```

```python
sim = Simulation(threads=10,folder=Path('../Lattice_Files/02_Aperture_Lattice/'))

line = sim.line
```

```python
opt = line.match(
    method='4d', 
    vary=[xt.VaryList(['kqd'], step=1e-8), xt.VaryList(['kqf'], step=1e-8)],
    targets = [xt.TargetSet(qx=4.47, qy=3.74, tol=1e-6, tag='tune')]
)

print(opt.get_knob_values())
```

```python
print(line.vars['kqf']._get_value())
print(line.vars['kqd']._get_value())
```

```python
line.vars['kqf'] = 0.0
line.vars['kqd'] = 0.0

for name in line.element_names:
    element = line[name]
    
    if isinstance(element, xt.Quadrupole):
        if 'qf' in name.lower():
            line.element_refs[name].k1 += line.vars['kqf']
            print(f"Linked {name} to kqf")
            
        elif 'qd' in name.lower():
            line.element_refs[name].k1 += line.vars['kqd']
            print(f"Linked {name} to kqd")

print(f"Active variables: {list(line.vars.keys())}")
```

```python
k_vars = [v for v in line.vars.keys() if v.startswith('k')]
print(k_vars)
```
